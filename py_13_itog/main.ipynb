```python
%matplotlib inline
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
churn_data = pd.read_csv('data\churn.csv')
churn_data = churn_data.drop(['RowNumber'],axis=1)
#churn_data['Exited']=churn_data['Exited'].apply(lambda x: 'exited' if x ==1 else 'not exited')
```

```python
fig_1 = px.pie(
    data_frame=churn_data,
    names = 'Exited',
    title = 'Соотношение ушедших клиентов и оставшихся'
)
fig_1.show()
```

По графику можно сделать вывод, что банк потерял 20.4% клиентов

```python
not_null_balance = churn_data[churn_data['Balance']>2500]
fig_2 = px.histogram(
    data_frame=not_null_balance,
    x = 'Balance'
)
fig_2.show()
```

```python
fig_3 = px.histogram(
    data_frame=churn_data,x = 'Balance', color ='Exited')
fig_3.show()
```

```python
fig_4 = px.box(
    data_frame=churn_data,x = 'Age', color ='Exited')
fig_4.show()
```

```python
fig = plt.figure(figsize=(15,15))
sc_plot = sns.scatterplot(x=churn_data['EstimatedSalary'],y=churn_data['CreditScore'],hue=churn_data['Exited'],sizes=1)
```
