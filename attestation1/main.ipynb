```python
def apply_discounts(products, stocks):
    for i in stocks:
        n = int(stocks[i][:-1])/100
        stocks[i] = n
    for j in products:
        if j in stocks.keys():
            products[j]=round(products[j] - products[j]*stocks[j],2)
    return products
```
