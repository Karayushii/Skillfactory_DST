```python
def quater(time):
    if 0<= time <=6:
        return 'night'
    elif 6< time <=12:
        return 'morning'
    elif 12< time <=18:
        return 'day'
    else:
        return 'evening'
```

```python
ct_bike['time_of_day']=ct_bike['starttime'].dt.hour.apply(quater)
```

```python
ct_bike[ct_bike['time_of_day']=='day']['time_of_day'].count()/ct_bike[ct_bike['time_of_day']=='night']['time_of_day'].count()
```
