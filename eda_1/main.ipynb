```python
def nan_percent(column_name):
    return round(100*(1-(df[column_name].count()/
        len(df[column_name]))))
```

```python
my_dict = dict()
for i in df.columns:
    my_dict[i]=nan_percent(i)
my_dict
```

```python
for i in df.columns:
    if nan_percent(i)>30:
        df = df.drop(i,axis=1)
df
```
