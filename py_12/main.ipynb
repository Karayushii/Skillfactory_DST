```python
melb_df = pd.read_csv('data\melb_data_fe.csv')
```

```python
def cat_cat(x):
    if x.nunique()<150 and x.name not in ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'] :
        return x.astype('category')
    return x
    
        
    
```

```python
melb_df['Date'].name
```

```python
melb_df = melb_df.apply(cat_cat)
```

```python
melb_df.info()
```

```python
ratings1 = pd.read_csv('data\\ratings1.csv')
ratings2 = pd.read_csv('data\\ratings2.csv')
ratings1['userId'].nunique()
```

```python
dates = pd.read_csv('data\dates.csv')
display(dates)
```

```python
dates['date']=pd.to_datetime(dates['date']) 
```

```python
dates['date'].dt.year.value_counts()
```

```python
ratings = pd.concat(
    [ratings1,ratings2],
    ignore_index=True)
display(ratings)
```

```python
print('Число строк в таблице ratings', ratings.shape[0])
print('Число строк в таблице ratings', dates.shape[0])
print(ratings.shape[0]==dates.shape[0])
```

```python
display(ratings1.tail(1))
display(ratings2.head(1))
```

```python
ratings = ratings.drop_duplicates(ignore_index=True)
display('Число строк в таблице ratings', ratings.shape[0])
```

```python
ratings_dates = pd.concat([ratings,dates], axis = 1)
display(ratings_dates.tail(7))
```

```python
movies = pd.read_csv('data\movies.csv')
```

```python
joined = ratings_dates.join(
    movies.set_index('movieId'),
    on='movieId',
    how = 'left'
)
display(joined.head())
```

```python
merged = ratings_dates.merge(
    movies,
    on='movieId',
    how='left'
)
display(merged.head())
```

```python
print('Число строк в таблице rating_dates',ratings_dates.shape[0])
print('Число строк в таблице merged',merged.shape[0])
```
